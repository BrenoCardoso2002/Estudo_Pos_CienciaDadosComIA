```python
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# --- CORREÇÃO DO ERRO ---
# Baixando os recursos necessários (Adicionado o punkt_tab)
nltk.download('punkt')
nltk.download('punkt_tab') # Esta linha resolve o LookupError
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

class ProcessadorTexto:
    def __init__(self):
        # Definimos o dicionário de palavras que serão removidas (artigos, preposições, etc)
        self.stop_words = set(stopwords.words('portuguese'))
        # O Lemmatizer reduz a palavra à sua forma base (ex: "estudando" vira "estudar")
        self.lemmatizer = WordNetLemmatizer()

    def limpar_texto(self, texto):
        # 1. Normalização: Tudo em minúsculo para o computador entender que "Olá" e "olá" são iguais
        texto = texto.lower()

        # 2. Limpeza de ruído: Remove links, menções e hashtags usando Expressões Regulares (Regex)
        texto = re.sub(r"http\S+|www\S+|@\S+|#\S+", '', texto)

        # 3. Pontuação: Remove símbolos que não agregam valor semântico (., ! ?)
        texto = texto.translate(str.maketrans('', '', string.punctuation))

        # 4. Tokenização: Divide a frase em uma lista de palavras individuais
        tokens = word_tokenize(texto)

        # 5. Filtragem e Lemmatização:
        # Removemos as stop words e aplicamos a raiz da palavra
        tokens_limpos = [self.lemmatizer.lemmatize(w) for w in tokens if w not in self.stop_words]

        # Retorna o texto "reconstruído" e limpo
        return " ".join(tokens_limpos)

# --- TESTE DO PROJETO 1 ---
texto_sujo = "Olá! Confira meu portfólio em http://github.com #IA. @Breno está estudando NLP na USJT!"

processador = ProcessadorTexto()
resultado = processador.limpar_texto(texto_sujo)

print("=== RESULTADO DO PROJETO 1 ===")
print(f"Original: {texto_sujo}")
print(f"Limpo:    {resultado}")
```
