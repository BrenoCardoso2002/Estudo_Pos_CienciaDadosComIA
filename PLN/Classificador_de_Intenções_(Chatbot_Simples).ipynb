```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# 1. Expandindo a base de conhecimento (Dataset)
# Quanto mais exemplos, mais "inteligente" o modelo fica.
dados = [
    # Intenção: Financeiro
    ("Como está meu saldo?", "financeiro"),
    ("Quero pagar uma conta", "financeiro"),
    ("Extrato bancário da semana", "financeiro"),
    ("Qual o valor da fatura?", "financeiro"),

    # Intenção: Suporte Técnico
    ("Meu app está travando", "suporte"),
    ("Não consigo fazer login", "suporte"),
    ("Preciso alterar minha senha", "suporte"), # <-- Adicionamos aqui!
    ("Erro ao abrir a página", "suporte"),
    ("Esqueci minha credencial de acesso", "suporte"),

    # Intenção: Saudação
    ("Obrigado pela ajuda", "saudacao"),
    ("Bom dia, como vai?", "saudacao"),
    ("Olá, gostaria de falar com alguém", "saudacao")
]

X_treino, y_treino = zip(*dados)

# 2. Criando o modelo
# O TfidfVectorizer transforma as palavras em números (vetores).
# O MultinomialNB é o classificador que aprende as probabilidades.
modelo_intencoes = make_pipeline(TfidfVectorizer(), MultinomialNB())

# 3. Treinando o modelo com os novos dados
modelo_intencoes.fit(X_treino, y_treino)

# --- TESTE DE FOGO ---
testes = [
    "Preciso alterar minha senha",
    "Quanto eu tenho na conta?",
    "O sistema parou de funcionar",
]

print("=== VALIDAÇÃO DE INTENÇÕES ===")
for frase in testes:
    predicao = modelo_intencoes.predict([frase])
    print(f"Frase: '{frase}' -> Intenção: {predicao[0]}")
```
